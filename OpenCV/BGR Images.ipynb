### Image Color Spaces And Conversion

```python
img = cv2.imread(r"C:\Users\hp\OneDrive\Pictures\wp7509160.jpg")

# convert the color to hsv
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

plt.imshow(hsv)
plt.axis('off')
plt.show()
```

```python
img = cv2.imread(r"C:\Users\hp\OneDrive\Pictures\wp7509160.jpg")

# convert the color to hsv
hLS = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)

plt.imshow(hLS)
plt.axis('off')
plt.show()
```

```python
img = cv2.imread(r"C:\Users\hp\OneDrive\Pictures\wp7509160.jpg")

# convert the color to hsv
YUP = cv2.cvtColor(img, cv2.COLOR_BGR2YUV)

plt.imshow(YUP)
plt.axis('off')
plt.show()
```

```python
img = cv2.imread(r"C:\Users\hp\OneDrive\Pictures\wp7509160.jpg")

# convert the color to hsv
RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(RGB)
plt.axis('off')
plt.show()
```

```python
#Access pixel values and modify them
pixel_value = img[100, 50]  # Accessing pixel value at (50, 100)
print("Pixel value before modification:", pixel_value)
img[100, 50] = [255, 255, 255]  # Modifying pixel value to white
pixel_value_modified = img[100, 50]
print("Pixel value after modification:", pixel_value_modified)
```

```python
# Access image properties
height, width, channels = img.shape
print("Image properties - Height:", height, "Width:", width, "Channels:", channels)
```

### Region Of Interest

```python
# Set a Region of Interest (ROI)
roi = img[50:1505, 100:1300]  # Selecting a region in the image
plt.imshow(roi)
plt.axis('off')
plt.show()
```

### SPLIT AND MERGE THE IMAGE CHANNEL

```python
B, G, R = cv2.split(image)
plt.imshow(B)
plt.axis('off')
plt.show()
```

```python
plt.imshow(G)
plt.axis('off')
plt.show()
```

```python
plt.imshow(R)
plt.axis('off')
plt.show()
```

```python
# Merge Image
merged_image = cv2.merge([B, G, R])
plt.imshow(merged_image)
plt.axis('off')
plt.show()
```

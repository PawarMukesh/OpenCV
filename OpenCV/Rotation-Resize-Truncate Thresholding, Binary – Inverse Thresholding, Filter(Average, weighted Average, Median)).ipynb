# 1. Import Library

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

# 2. Read Image


```python
image = cv2.imread('/content/pexels-photo-1996333.jpeg')
```

# 3. Show Image


```python
cv2_imshow(image)
```

# 4. Print an Image


```python
print(image)
```

# 5. Find Height & Width


```python
height, width = image.shape[:2]
print("Height:", height)
print("Width:", width)
```

# 6. Rotate an Image with Angle 160


```python
# Define the angle
angle = 160

# Define the rotation matrix
rotation_matrix = cv2.getRotationMatrix2D((width / 2, height / 2), angle, 1)

# Apply the rotation to the image
rotated_image = cv2.warpAffine(image, rotation_matrix, (width, height))

# Show the image
cv2_imshow(rotated_image)
```

# 7. Resize The Image


```python
resized_image = cv2.resize(image, (width // 2, height // 2))
cv2_imshow(resized_image)
```

```python
# Check orignal shape and resized image shape
print("Original Image Shape",image.shape[:2])
print("Resized Image Shape", resized_image.shape[:2])
```

# 8. Apply Truncate Thresholding, Binary â€“ Inverse Thresholding


```python
# Convert image into gray scale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply thresholding
_, binary_inverse_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY_INV | cv2.THRESH_TRUNC)

# Display the image
cv2_imshow(binary_inverse_image)
```

# 9. Perform Power Law Transformation using inverse Gamma


```python
gamma = 0.5
gamma_corrected = np.uint8(np.power(gray_image / 255.0, gamma) * 255)
cv2_imshow(gamma_corrected)
```

# 10. Apply Average Filter


```python
average_filtered = cv2.blur(image, (5, 5))
cv2_imshow(average_filtered)
```

# 11. Apply Weighted-Average Filter


```python
weighted_average_filtered = cv2.GaussianBlur(image, (5, 5), 0)
cv2_imshow(weighted_average_filtered)
```

# 12. Apply Median Filter


```python
median_filtered = cv2.medianBlur(image, 5)
cv2_imshow(median_filtered)
```

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread(r"C:\Users\hp\OneDrive\Pictures\A_Tug_of_War_in_Old_Zulu_Land.jpg", cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
img.shape
```

```python
# Select ROI From image

roi = img[100:200, 150:250]

# Display the image
plt.imshow(roi)
plt.axis('off')
plt.show()
```

```python
# Split and merge the image
B, G, R = cv2.split(img)


# Display the blue channel
print("Blue Channel: ")
plt.imshow(B)
plt.axis('off')
plt.show()

```

```python
# Display the Green channel
print("Green Channel: ")
plt.imshow(G)
plt.axis('off')
plt.show()
```

```python
# Display the red channel
print("Red Channel: ")
plt.imshow(R)
plt.axis('off')
plt.show()
```

```python
# Merge image channel into one
merge = cv2.merge([B, G, R])
plt.imshow(merge)
plt.axis('off')
plt.show()
```

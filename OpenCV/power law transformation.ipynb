```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read image
img = cv2.imread(r"C:\Users\hp\OneDrive\Pictures\desktop-wallpaper-lonely-dark-dark-sad-alone.jpg")

# Power low transformation
def power_low(x, c_value, gamma_value):
    return np.uint8(c_value * (x**gamma_value))
```

```python
cvalue = 3
gamma = 1
power_low_transformation = power_low(img,cvalue,gamma)


# Display original and power low transformed images
plt.figure(figsize=(10,5))
plt.subplot(1,2,1),plt.imshow(img, cmap='gray'), plt.axis('off'), plt.title('Original Image'), plt.show()
plt.subplot(1,2,1),plt.imshow(power_low_transformation, cmap='gray'), plt.axis('off'), plt.title('Transfrom Image'), plt.show()

plt.show()
```

## Check Intensity Of Pixel Of An GrayScale Image Using Histogram

```python
# Read Image
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
image = cv2.imread(r"C:\Users\hp\OneDrive\Pictures\mahadev-image-hd.jpg")

# Convert into gray scal
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Display the pixel values using histogram
plt.hist(gray)
plt.show()
```

```python
# Display the original Image
plt.imshow(gray, cmap='gray')
plt.axis('off')
plt.show()
```

## Check Intensity Of Pixel Of An RGB Image Using Histogram

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

image = cv2.imread(r"C:\Users\hp\OneDrive\Pictures\mahadev-image-hd.jpg")

blue = image[:,:,0]
green = image[:,:,1]
red = image[:,:,2]

# Display the pixel values using histogram

plt.hist(blue)
plt.title("Blue Channel")
plt.show()

plt.hist(green)
plt.title("Green Channel")
plt.show()

plt.hist(red)
plt.title("Red Channel")
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread(r"C:\Users\hp\OneDrive\Pictures\images.jpeg")
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
plt.axis('off')
plt.show()
```

```python
# Get height and width of the image
print(f"Hight and Width and Channel of the Image: {rgb_img.shape}")
```

```python
# Rotation of image
rotate_180 = cv2.rotate(rgb_img, cv2.ROTATE_180)
rotate_90_clockwise = cv2.rotate(rgb_img, cv2.ROTATE_90_CLOCKWISE)
rotate_90_counterclockwise = cv2.rotate(rgb_img, cv2.ROTATE_90_COUNTERCLOCKWISE)


# Rotate 180
plt.imshow(rotate_180)
plt.axis('off')
plt.show()

# rotate_90_clockwise
plt.imshow(rotate_90_clockwise)
plt.axis('off')
plt.show()

# rotate_90_counterclockwise
plt.imshow(rotate_90_counterclockwise)
plt.axis('off')
plt.show()
```

```python
# Get coordinates of image
height, width, channel = rgb_img.shape

center_x = height // 2
center_y = width // 2

print(f"Center X: {center_x}")
print(f"Center Y: {center_y}")
```

```python
# Apply rotation of the image of any angle
angle = 65

# Rotation
rotation = cv2.getRotationMatrix2D((center_x,center_y), angle, scale=1)

# Wrap the image
rotate_img = cv2.warpAffine(rgb_img, rotation, (height,width))

# Display the rotate image
plt.imshow(rotate_img)
plt.axis('off')
plt.show()
```

```python
# Image Scaling and Resize the image

scale_height = 20
scale_width = 20

# Calculate new diamensions 
new_height = int(rgb_img.shape[0] * scale_height)
new_width = int(rgb_img.shape[1] * scale_width)

# Resize the image
resized_img = cv2.resize(rgb_img,(new_height, new_width))

# Display the image
print("Original Image:")
plt.imshow(rgb_img)
plt.axis('off')
plt.show()

print("Resized Image:")
plt.imshow(resized_img)
plt.axis('off')
plt.show()
```

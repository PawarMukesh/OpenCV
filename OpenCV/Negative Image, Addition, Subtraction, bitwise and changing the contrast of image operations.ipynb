```python
# Rotate An Image ( Angle--100,360,50) 

# Get height and width
def rotation(image, angles):
    images = cv2.imread(image)
    for angle in angles:
        height, width = img.shape[0:2]
        
    # Get the center of image
    center = (width//2, height//2)

    # Apply rotation matrix
    rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1)
    
    # Warp image
    rotate_image = cv2.warpAffine(images, rotation_matrix, (height,width))
    
    # Display the rotated image
    print("Rotated Image: \n")
    plt.imshow(rotate_image)
    plt.axis('off')
    plt.show()



```

```python
img1 = 'bit im 2.png'
ang = [100,360,50]
rotation(img1,ang)
```

```python
# Image Scaling ( Resizing) 
def scale_resize(img):
    pic = cv2.imread(img)
    
    scale_height = int(input("Enter the scale height: "))
    scale_width = int(input("Enter the scale width: "))
    
    new_height = int(pic.shape[0] * scale_height)
    new_width = int(pic.shape[1] * scale_width)
    
    
    # Resize the image
    resize = cv2.resize(pic, (new_height,new_width))
    
    # Display the image
    plt.imshow(resize)
    plt.axis('off')
    plt.show()
```

```python
img2 = 'bit im 2.png'
scale_resize(img2)
```

```python
""" 8. How to Generate a Negative Image in Python using OpenCV   
        1. Step 1 — Importing libraries.  
         2. Step 2 — Let’s read the image.  
         3. Step 3 Convert the image to RGB  
         4. Step 4— Let's generate the negative image. """
```

```python
def negative(img):
    picture = cv2.imread(img)
    
    negative_image = 255 - picture
    
    print("Original Image: ")
    plt.imshow(picture)
    plt.axis('off')
    plt.show()
    
    print("Negative Image: ")
    plt.imshow(negative_image)
    plt.axis('off')
    plt.show()
```

```python
negative(img1)
```

```python
""" 9.Show each channel as a colored image  
        1. Binary Thresholding cv2.THRESH_BINARY  
        2. Truncate Thresholding cv2.THRESH_TRUNC   
        3. Zero Thresholding cv2.THRESH_TOZERO  """
```

```python
read = cv2.imread(r"C:\Users\hp\OneDrive\Pictures\8105706.jpg")
# Split the image into its color channels
b, g, r = cv2.split(read)

# Apply Binary Thresholding
_, binary_threshold_b = cv2.threshold(b, 127, 255, cv2.THRESH_BINARY)
_, binary_threshold_g = cv2.threshold(g, 127, 255, cv2.THRESH_BINARY)
_, binary_threshold_r = cv2.threshold(r, 127, 255, cv2.THRESH_BINARY)

# Apply Truncate Thresholding
_, truncate_threshold_b = cv2.threshold(b, 127, 255, cv2.THRESH_TRUNC)
_, truncate_threshold_g = cv2.threshold(g, 127, 255, cv2.THRESH_TRUNC)
_, truncate_threshold_r = cv2.threshold(r, 127, 255, cv2.THRESH_TRUNC)

# Apply Zero Thresholding
_, zero_threshold_b = cv2.threshold(b, 127, 255, cv2.THRESH_TOZERO)
_, zero_threshold_g = cv2.threshold(g, 127, 255, cv2.THRESH_TOZERO)
_, zero_threshold_r = cv2.threshold(r, 127, 255, cv2.THRESH_TOZERO)

# Merge the thresholded channels back into a colored image
binary_threshold_image = cv2.merge([binary_threshold_b, binary_threshold_g, binary_threshold_r])
truncate_threshold_image = cv2.merge([truncate_threshold_b, truncate_threshold_g, truncate_threshold_r])
zero_threshold_image = cv2.merge([zero_threshold_b, zero_threshold_g, zero_threshold_r])


# Display the original and thresholded images
plt.figure(figsize=(10, 6))

plt.subplot(2, 2, 1), plt.imshow(cv2.cvtColor(read, cv2.COLOR_BGR2RGB)), plt.title('Original Image'), plt.axis('off')
plt.subplot(2, 2, 2), plt.imshow(cv2.cvtColor(binary_threshold_image, cv2.COLOR_BGR2RGB)), plt.title('Binary Thresholding'), plt.axis('off')
plt.subplot(2, 2, 3), plt.imshow(cv2.cvtColor(truncate_threshold_image, cv2.COLOR_BGR2RGB)), plt.title('Truncate Thresholding'), plt.axis('off')
plt.subplot(2, 2, 4), plt.imshow(cv2.cvtColor(zero_threshold_image, cv2.COLOR_BGR2RGB)), plt.title('Zero Thresholding'),plt.axis('off')

plt.show()
```

```python
# 10. Addition of Image 
#  1.  Read the images. 
#  2. Add the images    1.  cv2.add(img1, img2) , 2. addWeighted 

photo1 = cv2.imread(r"C:\Users\hp\OneDrive\Pictures\bit im 1.png")
photo2 = cv2.imread(r"C:\Users\hp\OneDrive\Pictures\bit im 2.png")

add = cv2.add(photo1,photo2)

plt.imshow(add)
plt.axis('off')
plt.show()
```

```python
# 11. Subtraction of Image 
subtract = cv2.subtract(photo1,photo2)

plt.imshow(subtract)
plt.axis('off')
plt.show()
```

```python
""" 12. Bitwise Operations  -  Bitwise operations are used in image manipulation and used for extracting essential parts in the image 
               1. AND       2.OR        3.XOR        4.NOT  
                  Syntax-  cv2.bitwise_and  
                           cv2.bitwise_or 
                           cv2.bitwise_xor 
                           cv2.bitwise_not  """
```

```python
# Bitwise AND, OR, XOR, NOT operations
result_and = cv2.bitwise_and(photo1, photo2)
result_or = cv2.bitwise_or(photo1, photo2)
result_xor = cv2.bitwise_xor(photo1, photo2)
result_not = cv2.bitwise_not(photo1)

# Display the images
plt.figure(figsize=(10, 6))

plt.subplot(2, 3, 1), plt.imshow(photo1, cmap='gray'), plt.title('Image 1'), plt.axis('off')
plt.subplot(2, 3, 2), plt.imshow(photo2, cmap='gray'), plt.title('Image 2'), plt.axis('off')
plt.subplot(2, 3, 3), plt.imshow(result_and, cmap='gray'), plt.title('AND'), plt.axis('off')
plt.subplot(2, 3, 4), plt.imshow(result_or, cmap='gray'), plt.title('OR'), plt.axis('off')
plt.subplot(2, 3, 5), plt.imshow(result_xor, cmap='gray'), plt.title('XOR'), plt.axis('off')
plt.subplot(2, 3, 6), plt.imshow(result_not, cmap='gray'), plt.title('NOT'), plt.axis('off')

plt.show()
```

```python
# 13. change the contrast and brightness of an image using OpenCV 
#             1. read the input image 
#             2. define the alpha and beta 
#             3.  call convertScaleAbs function cv2.convertScaleAbs(image, alpha, beta) 
#             4.   also use Method ------addWeight ed


# Define alpha and beta
alpha = 1.7
beta = 25

convert = cv2.convertScaleAbs(read, alpha=alpha, beta=beta)

wighted = cv2.addWeighted(read, alpha,np.zeros_like(read),0, beta)

plt.figure(figsize=(10,6), facecolor='white')
plt.subplot(1,2,1), plt.imshow(convert), plt.axis('off'), plt.title("Convert Scale Abs"), plt.show()
plt.subplot(1,2,2), plt.imshow(wighted), plt.axis('off'), plt.title("Add Weighted"), plt.show()

plt.show()
```

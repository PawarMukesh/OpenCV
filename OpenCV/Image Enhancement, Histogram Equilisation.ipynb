```python
import cv2
```

```python
img = cv2.imread(r"C:\Users\hp\OneDrive\Pictures\mahadev-image-hd.jpg")

# Convert into gray scale image 
gray_scale_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# equiliasation of image
equilized_img = cv2.equalizeHist(gray_scale_img)
cv2.imshow('Image',img)
cv2.imshow('Gray Scale Image',gray_scale_img)
cv2.imshow('Equilized Image', equilized_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Import necessary Library
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Visualise the image pixel of gray scale image and equlized image
plt.subplot(1,1,1)
hist = cv2.calcHist(gray_scale_img,[0],None,[255],[0,256])
plt.plot(hist, color='black')
plt.show()

plt.subplot(1,1,1)
equi_hist = cv2.calcHist(equilized_img,[0],None,[255],[0,256])
plt.plot(equi_hist, color='black')
plt.show()
```

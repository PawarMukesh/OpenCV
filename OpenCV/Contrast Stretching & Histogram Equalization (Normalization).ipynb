```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read the image
image = cv2.imread(r"C:\Users\hp\OneDrive\Pictures\futuristic-robot-artificial-intelligence-huminoid-ai-industrial-factory_31965-48268.webp")

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply contrast stretching
min_pixel_value = np.min(gray_image)
max_pixel_value = np.max(gray_image)
stretched_image = ((gray_image - min_pixel_value) / (max_pixel_value - min_pixel_value)) * 120

# Apply histogram equalization
equalized_image = cv2.equalizeHist(gray_image)

# Display the original, stretched, and equalized images
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1), plt.imshow(gray_image, cmap='gray'), plt.title('Original Image'), plt.axis('off')
plt.subplot(1, 3, 2), plt.imshow(stretched_image, cmap='gray'), plt.title('Contrast Stretched Image'), plt.axis('off')
plt.subplot(1, 3, 3), plt.imshow(equalized_image, cmap='gray'), plt.title('Equalized Image'), plt.axis('off')

plt.show()
```

```python
# plot the histogram of original image & Equilize Image
hist = cv2.calcHist([gray_image],[0],None, [255],[0,256])
plt.plot(hist,color='black')
plt.xlabel('Pixel Value')
plt.ylabel('Fequnacy')
plt.title("Gray Image Histogram")
plt.show()


equi_hist = cv2.calcHist([equalized_image],[0],None, [255],[0,256])
plt.plot(equi_hist,color='black')
plt.xlabel('Pixel Value')
plt.ylabel('Fequnacy')
plt.title("Equalized Image Histogram")
plt.show()
```

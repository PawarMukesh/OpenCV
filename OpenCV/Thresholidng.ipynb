```python
#  8. Rotate An Image ( Angle--120,160,30)
def rotate_image(image, angles):
    original_image = cv2.imread(image)
    
    for angle in angles:
        height, width = original_image.shape[:2]
        
        # Calculate rotation matrix
        center = (width//2, height//2)
        rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1)
        
        # Apply rotation to image
        roate_image = cv2.warpAffine(original_image, rotation_matrix, (width, height))
        
        # Display the rotated image
        print("Rotated Image: \n")
        plt.imshow(roate_image)
        plt.axis('off')
        plt.show()
```

```python
img = 'img.jpg'
rotation_angle = [120,160,30]
rotate_image(img, rotation_angle)
```

```python
#  9. Image Scaling ( Resizing)

def resize_image(img):
    
    pic = cv2.imread(img)
    
    # Scaling Factor for the given image
    scale_height = int(input("Enter the height scale factor: "))
    scale_width = int(input("Enter the width scale factor: "))
    
    # Calculate new dimension
    new_height = int(pic.shape[0] * scale_height)
    new_width = int(pic.shape[1] * scale_width)
    
    # Resize the give image
    resize = cv2.resize(pic, (new_height, new_width))
    
    # Display the image
    plt.imshow(resize)
    plt.axis('off')
    plt.show()
```

```python
picture = 'img.jpg'
resize_image(picture)
```

```python
# 10. How to Generate a Negative Image in Python using OpenCV
#  1. Step 1 — Importing libraries.
#  2. Step 2 — Let’s read the image.
#  3. Step 3 Convert the image to RGB
#  4. Step 4— Let's generate the negative image.


img = cv2.imread('img.jpg')

# Convert image into negative
negative_img = 255 - img

# Display original imge
print("Original Image: \n ")
plt.imshow(img)
plt.axis('off')
plt.show()



print("Negative Image: \n")
plt.imshow(negative_img)
plt.axis('off')
plt.show()
```

```python
# 11.Show each channel as a colored image
#  1. Binary Thresholding cv2.THRESH_BINARY
#  2. Binary – Inverse Thresholding cv2.THRESH_BINARY_INV
#  3. Truncate Thresholding cv2.THRESH_TRUNC
#  4. Zero Thresholding cv2.THRESH_TOZERO
#  5. Inverse Thresholding to Zero cv2.THRESH_TOZERO_INV 
 
 
 
# Read the image
image = cv2.imread('img.jpg')

# Convert the image to RGB format
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Split the image channel
b, g, r = cv2.split(image_rgb)

# Thresholding techniques
threshold_types = [cv2.THRESH_BINARY, cv2.THRESH_BINARY_INV, cv2.THRESH_TRUNC, cv2.THRESH_TOZERO, cv2.THRESH_TOZERO_INV]
thresholded_images = []

for threshold_type in threshold_types:
    _, thresholded_r = cv2.threshold(r, 128, 255, threshold_type)
    _, thresholded_g = cv2.threshold(g, 128, 255, threshold_type)
    _, thresholded_b = cv2.threshold(b, 128, 255, threshold_type)

    # Merge thresholded channels back to form a colored image
    thresholded_image = cv2.merge([thresholded_b, thresholded_g, thresholded_r])
    thresholded_images.append(thresholded_image)

# Display the original and thresholded images side by side
fig, axes = plt.subplots(2, len(threshold_types) + 1, figsize=(15, 5))

axes[0, 0].imshow(image_rgb)
axes[0, 0].set_title('Original Image')

for i in range(len(threshold_types)):
    axes[0, i + 1].imshow(thresholded_images[i])
    axes[0, i + 1].set_title(f'Threshold Type {i + 1}')

# Display individual channels
axes[1, 0].imshow(r, cmap='gray')
axes[1, 0].set_title('Red Channel')

for i in range(len(threshold_types)):
    _, thresholded_r = cv2.threshold(r, 128, 255, threshold_types[i])
    axes[1, i + 1].imshow(thresholded_r, cmap='gray')
    axes[1, i + 1].set_title(f'Threshold Type {i + 1}')

for ax in axes.flatten():
    ax.axis('off')

plt.show()
```

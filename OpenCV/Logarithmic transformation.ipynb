```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
# Read the image
img = cv2.imread(r"C:\Users\hp\OneDrive\Pictures\desktop-wallpaper-lonely-dark-dark-sad-alone.jpg")

# Logarithmic transformation function
def log_transformation(x, c_value):
    return np.uint8(c_value * np.log1p(x))
```

```python
# apply logarithimc transformation
c_value = 60 #adjust this parameter for scaling
transformed_img = log_transformation(img, c_value)

# show the image
plt.figure(figsize=(5,5))
plt.subplot(1,2,1),plt.imshow(img,cmap='gray'),plt.axis('off'),plt.title('Original Image'),plt.show()
plt.subplot(1,2,2),plt.imshow(transformed_img,cmap='gray'),plt.axis('off'),plt.title('Logarithmic Image'),plt.show()

plt.show()
```
